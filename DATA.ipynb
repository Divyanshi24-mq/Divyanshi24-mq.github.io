```python
# Example cleaning function
def clean_numeric_columns(df):
    # Convert columns to numeric where possible, coerce errors to NaN
    df = df.apply(pd.to_numeric, errors='coerce')
    # Drop rows or columns with NaN values if necessary
    df = df.dropna()
    return df

# Apply cleaning to each DataFrame where necessary
df_income_clean = clean_numeric_columns(df_income)
df_housing_affor_clean = clean_numeric_columns(df_housing_affor)

```

```python
# Correlation function for cleaned dataframes
def calculate_correlation(df, name):
    print(f"\nCorrelation Matrix for {name}:")
    print(df.corr())

# Apply the correlation function to each cleaned DataFrame
calculate_correlation(df_income_clean, "Employee Median Income")
calculate_correlation(df_housing_affor_clean, "Housing Affordability")
# Continue for other cleaned DataFrames
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = [
    "Pensioner and beneficiary LCI (PBLCI)",
    "Employee LCI",
    "Age pensioner LCI",
    "Other government transfer recipient LCI",
    "Self-funded retiree LCI",
    "Consumer Price Index (CPI)"
]
quarterly_change = [1.3, 1.3, 1.2, 1.4, 1.2, 1.0]
annual_change = [4.1, 6.2, 3.7, 4.6, 3.8, 3.8]

# Setting up the positions and width for the bars
y = np.arange(len(categories))
bar_width = 0.4

# Plotting
fig, ax = plt.subplots(figsize=(10, 7))
bars1 = ax.barh(y - bar_width/2, quarterly_change, height=bar_width, label='Change from previous quarter (%)', color='skyblue')
bars2 = ax.barh(y + bar_width/2, annual_change, height=bar_width, label='Annual change (%)', color='steelblue')

# Labels and Titles
ax.set_yticks(y)
ax.set_yticklabels(categories)
ax.set_xlabel("Percentage change (%)")
ax.set_title("Selected Living Cost Indexes (LCIs) by household type and CPI, Australia")

# Add values to bars
for bar in bars1:
    ax.text(bar.get_width() + 0.1, bar.get_y() + bar.get_height()/2, f'{bar.get_width()}%', va='center')
for bar in bars2:
    ax.text(bar.get_width() + 0.1, bar.get_y() + bar.get_height()/2, f'{bar.get_width()}%', va='center')

# Legend and layout adjustment
ax.legend(loc="upper right")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Sample data
data = {
    'Category': [
        'Improvised dwellings, tents, or sleeping out',
        'Supported accommodation for the homeless',
        'Staying temporarily with other households',
        'Living in boarding houses',
        'Other temporary lodgings',
        "Living in 'severely' crowded dwellings"
    ],
    'Males': [5, 20, 10, 15, 8, 42],
    'Females': [6, 25, 15, 12, 9, 33]
}

# Create DataFrame and reshape for Plotly
df = pd.DataFrame(data)
df_melted = df.melt(id_vars='Category', var_name='Sex', value_name='Percentage')

# Custom color palette for each category
colors = ['#2A9D8F', '#E9C46A', '#F4A261', '#E76F51', '#264653', '#A8DADC']

# Create figure
fig = go.Figure()

# Add bars for each category in a stacked bar format with smooth color gradients
for idx, category in enumerate(df['Category']):
    fig.add_trace(go.Bar(
        x=df_melted[df_melted['Category'] == category]['Sex'],
        y=df_melted[df_melted['Category'] == category]['Percentage'],
        name=category,
        marker=dict(color=colors[idx], line=dict(width=0.5, color='white')),
        hoverinfo="x+y+name",
        text=df_melted[df_melted['Category'] == category]['Percentage'],
        textposition='inside'
    ))

# Customize layout to enhance visual appeal
fig.update_layout(
    barmode='stack',
    title=dict(
        text="Youth Homelessness by Type and Gender, Australia, 2021",
        font=dict(size=24, family="Arial", color="#333"),
        x=0.5
    ),
    xaxis=dict(
        title="Gender",
        titlefont=dict(size=18, family="Arial", color="#333"),
        tickfont=dict(size=14, family="Arial", color="#555")
    ),
    yaxis=dict(
        title="Percentage (%)",
        titlefont=dict(size=18, family="Arial", color="#333"),
        tickfont=dict(size=14, family="Arial", color="#555"),
        range=[0, 100],
        gridcolor="rgba(200,200,200,0.3)"
    ),
    legend=dict(
        title="Homelessness Type",
        font=dict(size=12, color="#555"),
        orientation="h",
        yanchor="bottom",
        y=-0.2,
        xanchor="center",
        x=0.5,
        bgcolor="rgba(255,255,255,0.7)",
        bordercolor="rgba(0,0,0,0.1)",
        borderwidth=1
    ),
    plot_bgcolor="white",
    margin=dict(l=40, r=40, t=80, b=100)
)

# Update hover template for a clean look
fig.update_traces(
    hovertemplate="<b>%{x}</b><br>Category: %{name}<br>Percentage: %{y}%<extra></extra>"
)

# Enhance interactivity and smooth transitions
fig.update_layout(transition_duration=500)

# Display the figure
fig.show()
```
